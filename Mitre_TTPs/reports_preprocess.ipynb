```python
# Rename .pdf files & save file information(hash, name, url)

import os
import csv
import hashlib

file_hash_name_url_list = [] # {file_hash, original_file_name, url}

path = r"C:\Users\xiaowan\Documents\GitHub\AttacKG\data\cti\pdf"
file_list = os.listdir(path)
for file in file_list:
    file_hash = 0
    with open(os.path.join(path, file), 'rb') as cti_file:
        file_hash = hashlib.md5(cti_file.read()).hexdigest()

    match = 0
    for url in url_list:
        if url.find(file) != -1:
            match += 1
            file_hash_name_url_list.append([file_hash, file, url])
            os.rename(os.path.join(path, file), os.path.join(path, file_hash+'.pdf'))

print(file_hash_name_url_list)

with open(os.path.join(path, "file_hash_name_url_list.csv"), 'w+') as csv_file:
    write = csv.writer(csv_file)
    write.writerows(file_hash_name_url_list)
```
